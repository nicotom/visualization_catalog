## Seaborn

```python
# Import library
import seaborn as sns
```

```python
# Load dataset
df = sns.load_dataset('iris')
```

```python
# Make default histogram of sepal length
ax = sns.distplot( df["sepal_length"] )
```

```python
# Control the number of bins
ax = sns.distplot( df["sepal_length"], bins=20 )
```

```python
# Histogram only
ax = sns.distplot( df["sepal_length"], bins=10 , kde=False)
```

```python
# Gaussian kernel density estimate only
ax = sns.distplot( df["sepal_length"], bins=10 , kde=True, hist=False)
```
