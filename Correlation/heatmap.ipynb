## Seaborn

```python
# library
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Create a dataset (fake)
df = pd.DataFrame(np.random.random((5,5)), columns=["a","b","c","d","e"])
```

```python
# Default heatmap: just a visualization of this square matrix
p1 = sns.heatmap(df)
```

```python
# Annotate each cell with the value
p2 = sns.heatmap(df, cmap="coolwarm", annot=True, annot_kws={"size": 8})
```

```python
# Create a dataset (fake)
df = pd.DataFrame(np.random.random((100,5)), columns=["a","b","c","d","e"])
```

```python
# Calculate correlation between each pair of variable
corr_matrix=df.corr()

# Can be great to plot only a half matrix
mask = np.zeros_like(corr_matrix)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    ax = sns.heatmap(corr_matrix, mask=mask, square=True, cmap=sns.diverging_palette(145, 280, s=85, l=25, n=7))
_ = ax.set_title('Correlation Half-Matrix')
```
